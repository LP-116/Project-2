```python
suburbs_data = suburbs_data.rename(columns={"Suburb/Town Name": "suburb", "LGA": "lga", "Lat":"latitude", "Lng": "longitude"})
suburbs_data
```

```python
suburbs_data = suburbs_data.dropna(how="any")
suburbs_data
```

```python
crime_data = "Criminal_incidents.csv"
```

```python
crime_data = pd.read_csv(crime_data)
```

```python
crime_data.head()
```

```python
crime_df = crime_data.drop(columns=["Year ending", "Offence Subgroup", "Unnamed: 9"])
```

```python
crime_df.index.name = "index_no"
crime_df
```

```python
crime_df = crime_df.rename(columns={"Year":"year","Suburb/Town Name": "suburb", "Postcode":"postcode", "Local Government Area": "lga", "Offence Division": "offence_div",
                                   "Offence Subdivision":"offence_sub_div", "Incidents Recorded": "incidents"})
```

```python
crime_df = crime_df[["year", "suburb", "lga", "postcode", "offence_div", "offence_sub_div", "incidents"]]
```

```python
crime_df.head()
```

```python
crime_df = crime_df.dropna(how="any")
crime_df.head()
```

```python
crime_df.replace(',','', regex=True, inplace=True)
crime_df[["year", "postcode", "incidents"]] = crime_df[["year", "postcode", "incidents"]].apply(pd.to_numeric)
```

```python
lyons_loc = crime_df.loc[crime_df["suburb"] == "Lyons"].index
lyons_loc
```

```python
crime_df = crime_df.drop(lyons_loc)
```

```python
crime_df.to_csv("crime.csv", encoding="utf-8", index=False, header=True)
suburbs_data.to_csv("suburb-final.csv", encoding="utf-8", index=False, header=True)
```

```python
first = pd.read_csv("crime.csv")
second = pd.read_csv("suburb-final.csv")
```

```python
combined_df = pd.merge(first, second, on="suburb", how="left")
combined_df
```

```python
combined_df = combined_df.drop(columns=["lga_y"])
combined_df = combined_df.rename(columns={"lga_x":"lga"})
```

```python
combined_df = combined_df[["year", "suburb", "latitude", "longitude", "lga", "postcode", "offence_div", "offence_sub_div", "incidents"]]
combined_df
```

```python
combined_df.index.name = "index_no"
combined_df
```

```python
combined_df.to_sql(name='crime', con=engine, if_exists='append', index=True)
```
